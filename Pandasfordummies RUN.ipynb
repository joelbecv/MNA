```python
adult_df.education.loc[10002]#muestra los resultados de la ediucación de la tercer fila (una de varias opciones)
```

```python
adult_df["education"].iloc[2]#muestra los resultados de la ediucación de la tercer fila (una de varias opciones)
```

```python
adult_df.at[10002,"education"]  #Accessing values with .at[] is my favorite for two reasons. First, it is much neater and more straightforward. Second, you can treat the DataFrame like a matrix as it is one, at least visuall
```

## Series Access

```python
row_series = adult_df.loc[10002] #accesa a la fila 10002 y la almacena en una variable llamada row_series
print(row_series.loc["education"]) #se indica la columna que quiero que me regres con el label de la columna
print(row_series.iloc[3]) #se indica la columna con la ubicación indexada de la columna
print(row_series["education"]) #método 3  para lo mismo
print(row_series.education)#método 4 para lo mismo
```

```python
column_series = adult_df.education  #accesa a la columna education y la almacena en una variable llamada column_series
```

```python
print(column_series.loc[10002])
print(column_series.loc[10002])
print(column_series.iloc[2])
print(column_series[10002])
```
